## NumPy Basics
NumPy is used for numerical operations in Python.

**Why NumPy?**
- Faster and more efficient than Python lists for numerical computations
- Supports powerful operations like broadcasting and matrix manipulation

````python
import numpy as np

# NumPy arrays are more efficient than Python lists for numerical operations
arr = np.array([1, 2, 3, 4, 5])
print("NumPy Array:", arr)
print("Mean:", np.mean(arr))
print("Standard Deviation:", np.std(arr))

# Element-wise operations
arr2 = arr * 2
print("Doubled Array:", arr2)

# Matrix operations
matrix = np.array([[1, 2], [3, 4]])
print("Matrix:", matrix)
print("Transpose:", matrix.T)
````

## **Data Cleaning, Wrangling & Visualization**

### üß≠ Objective:
Learn to clean messy data and explore insights visually using Pandas, Matplotlib, and Seaborn.

### ‚úÖ 1. Handling Missing Data

````python
import pandas as pd

# Load Titanic dataset again
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

# Check for missing values
df.isnull().sum()
````

````python
# Drop rows with missing values
df_dropped = df.dropna()

# Fill missing values (e.g., Age)
df['Age_filled'] = df['Age'].fillna(df['Age'].mean())
````

````python
# Fill missing 'Embarked' with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
````

### ‚úÖ 2. Removing Duplicates and Outliers

````python
# Check for duplicates
print("Duplicates:", df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()

# Detecting outliers using IQR
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['Fare'] < Q1 - 1.5 * IQR) | (df['Fare'] > Q3 + 1.5 * IQR)]
outliers.head()
````

### ‚úÖ 3. Filtering and Conditional Selection

````python
# Passengers older than 60
df[df['Age'] > 60]
````

````python
# Female passengers in first class
df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]
````

### üîÅ Before and After: Handling Missing Values

````python
# BEFORE: Check missing values
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df.isnull().sum()
````

````python
# AFTER: Fill missing Age and Embarked
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df.isnull().sum()
````

### üîÅ Before and After: Handling Duplicates

````python
# BEFORE: Check for duplicates
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
print("Duplicates before:", df.duplicated().sum())
````

````python
# AFTER: Drop duplicates
df = df.drop_duplicates()
print("Duplicates after:", df.duplicated().sum())
````

### üîÅ Before and After: Detecting Outliers

````python
# BEFORE: Show Fare summary stats
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df['Fare'].describe()
````

````python
# AFTER: Remove outliers using IQR
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['Fare'] >= Q1 - 1.5 * IQR) & (df['Fare'] <= Q3 + 1.5 * IQR)]
df['Fare'].describe()
````

### üìä 4. Data Visualization with Seaborn & Matplotlib

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram of Age
sns.histplot(df['Age'].dropna(), kde=True)
plt.title('Age Distribution')
plt.show()
````

````python
# Boxplot: Fare by Class
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare by Passenger Class')
plt.show()
````

````python
# Pairplot
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived')
plt.suptitle('Pairplot - Titanic Data', y=1.02)
plt.show()
````

````python
# Countplot of survival
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
````
